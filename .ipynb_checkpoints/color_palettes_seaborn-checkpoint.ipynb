```python
import seaborn as sns
```

```python
#Seaborn default style
fig, axes = plt.subplots(figsize=(6,4))
for inter in range(0, 3500, 500):
    axes.plot(x,y+inter, lw=2.5)
axes.set_title("Using default Seaborn Style")
```

```python
#Modified the original scheme plotting function to add names 
#for the palette name.
def palplot(pal, ax, size=1):
    """Plot the values in a color palette as a horizontal array.

    Parameters
    ----------
    pal : sequence of matplotlib colors
        colors, i.e. as returned by seaborn.color_palette()
    size :
        scaling factor for size of plot

    """
    n = len(pal)
    ax.imshow(np.arange(n).reshape(1, n),
              cmap=mpl.colors.ListedColormap(list(pal)),
              interpolation="nearest", aspect="auto")
    ax.set_xticks(np.arange(n) - .5)
    ax.set_yticks([-.5, .5])
    ax.set_xticklabels([])
    ax.set_yticklabels([])
```

```python
def palette_joint_plotter(list_of_palettes, axes, let_color="white"):
    for pal, ax in zip(list_of_palettes, axes.flatten()):
        current_palette = sns.color_palette(pal)
        palplot(current_palette, ax)
        ax.text(-0.4,0.1, pal, fontsize=15, color=let_color, fontweight='bold')

#Plotting all available palletes
sns.set_style("darkgrid")

fig, axes = plt.subplots(3,2, figsize=(12,3))
available_palettes = ["deep", "muted", "bright", "pastel", "dark", "colorblind"]
palette_joint_plotter(available_palettes, axes)
```

```python
#x,y for plotting puposes
x = np.arange(-10,10, 0.2)
y = x**3
```

```python
#Using hls for selecting colors
current_palette_4 = sns.color_palette("hls", 4)
sns.set_palette(current_palette_4)
sns.palplot(current_palette_4)
fig, axes = plt.subplots(figsize=(6,4))
for inter in range(0, 3500, 500):
    axes.plot(x,y+inter, lw=2.5)
axes.set_title("hls 4 color palette")
current_palette_7 = sns.color_palette("hls", 7)
sns.set_palette(current_palette_7)
sns.palplot(current_palette_7)
fig, axes = plt.subplots(figsize=(6,4))
for inter in range(0, 3500, 500):
    axes.plot(x,y+inter, lw=2.5)
axes.set_title("hls 7 color palette")
```

```python
#Plotting qualitative brewer palettes
fig, axes = plt.subplots(4,2, figsize=(12,4))
color_brew_qual = ["Set1", "Set2", "Set3", "Paired", "Accent", "Pastel1", "Pastel2", "Dark2"]
palette_joint_plotter(color_brew_qual, axes)
```

```python
#Setting qualitative data with color brew
color_brewer = sns.choose_colorbrewer_palette("qualitative")
```

```python
#plotting with selected brewer value
current_palette = sns.color_palette("Paired")
sns.set_palette(current_palette)
fig, axes = plt.subplots()
x = np.arange(0,10, 3)
for intercept in range(2, 10, 2):
    y = intercept + x 
    axes.plot(x,y, lw=2.5)
    y = intercept  - x 
    axes.plot(x,y, lw=2.5)
axes.set_title("Plotting with Color Brew")
```

```python
#Setting own color palette using named colors
#defining list
col_list = ["warm grey", "gunmetal", "dusky blue", "cool blue", "deep teal", "viridian", "twilight blue"]
sns.palplot(sns.xkcd_palette(col_list))
```

```python
#Plotting with color list defined
col_list_palette = sns.xkcd_palette(col_list)
sns.set_palette(col_list_palette)
fig, axes = plt.subplots(figsize=(6,4))
x_1 = np.arange(0,11, 1)
x_2 = np.arange(-10, 1, 1)
for intercept in range(2, 10, 2):
    y = intercept - x_2 
    axes.plot(x_2,y, lw=2.5)
    y = intercept  + x_2 
    axes.plot(x_2,y, lw=2.5)
axes.set_title("Using defined color palette")
```

```python
#Plotting all the possibilities of Color Brew for sequential colors
fig, axes = plt.subplots(9,2, figsize=(12,8))
color_brew_seq = ["Greys", "Reds", "Greens", "Blues", "Oranges", "Purples", 
                  "BuGn", "BuPu", "GnBu", "OrRd", "PuBu", "PuRd", "RdPu", "YlGn", 
                  "PuBuGn", "YlGnBu", "YlOrBr", "YlOrRd"]
palette_joint_plotter(color_brew_seq, axes, let_color="#CAADAC")
```

```python
#You can reverse the order of the color palette by adding the subscript "_r".
seq_col_brew = sns.color_palette("Blues_r", 4)
sns.set_palette(seq_col_brew)
sns.palplot(seq_col_brew)
fig, axes = plt.subplots(figsize=(6,4))
x = np.arange(-10,10,0.5)
for inter in range(0, 4000, 500):
    axes.plot(x,x**3+inter, lw=2.5)
axes.set_title("Brewer sequential reversed Blues palette")
```

```python
#Creating a variance covariance matrix. Note that this is only for illustration purposes.
#Note that final var cov matrix may not be positive definite.
mean = np.repeat(0,10)
emp_matrix = np.zeros([10,10])
for i in range(len(emp_matrix)):
    corrls = np.random.uniform(-1, 1, i)
    comp_zeros = np.repeat(0, 10-i)
    row_insert = np.concatenate((corrls, comp_zeros))
    emp_matrix[i] = row_insert
final_corr_mat = emp_matrix + emp_matrix.T
final_corr_mat
np.fill_diagonal(final_corr_mat, 1)
```

```python
#Plotting correlation matrix
plt.matshow(final_corr_mat, cmap=sns.diverging_palette(128, 240,as_cmap=True))
plt.xticks(np.arange(0,10))
plt.yticks(np.arange(0,10))
plt.title("Correlation Matrix")
```
